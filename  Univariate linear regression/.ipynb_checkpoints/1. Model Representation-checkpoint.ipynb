```python
#Linear Regression with one variable - univariate

#library which is used for ndarrays for example and images in RGB representation
import numpy as np
#library for visualisation
import matplotlib.pyplot as plt
```

```python
#X_train is the input variable (size in 1000 square feet) - input or features or variable or x
#y_train is the target (price in 1000s of dollars) - target or output or y
#input and output training set are separate
x_train = np.array([1.0,2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train= {x_train}")
print(f"y_train= {y_train}")
```

```python
#m - number of training examples. Numpy shape returns tuple for each dimension.

#it is one dimensional array so number of rows are 2
print(f"x_train.shape : {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
#Other option using len function
#maybe more appropriate
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
#All training examples
#number of X Y pair
for i in range(m):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
#Ploting the data
#show input and ouput pair like red x on the plot
plt.scatter(x_train, y_train, marker="x", c='r')
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of $)")
plt.xlabel("Price (1000 sqft)")
plt.show()
```

```python
#Model parameters
#linear regression model has two params w for slope and b for intercept
#expression is f(x)=w*x+b
w = 200
b = 100

print(f"w:{w}")
print(f"b:{b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      y (ndarray (m,)): target values
    """
    m = x.shape[0]
    #create an array of m elements with value 0
    f_wb = np.zeros(m)
    #update the elemenents - f_wb[i] will be prediction of y (or y hat)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb 
```

```python
#Plotting the output

#for INPUT training set get prediction of output value
lr_model = compute_model_output(x_train, w, b)

#Plot our model prediction - x training set and y hat (not y!!) --> line
plt.plot(x_train, lr_model, c='b', label='Model prediction')
#Add data points --> graph with two red x
plt.scatter(x_train, y_train, marker='x', c='r', label ='Actual values')

plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of $)")
plt.xlabel("Size (1000s sqft)")
plt.legend()

plt.show()
```

```python
#PREDICTION

#House with 1200 sqft. 
#same params
w = 200
b = 100
#one test example
x_i = 1.2

#get a prediction (not actual value) for y (that is y hat)
cost_1200sqft = w * x_i + b

print(f"${cost_1200sqft: .0f} thousand dollars")
```
